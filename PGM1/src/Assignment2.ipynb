```python
handle = Entrez.esearch(db='gene', term='PGM1')
record = Entrez.read(handle, validate=True)
print(pprint(record))
```

```python
list_of_genBank = record['IdList']
```

```python
list_of_genBank
```

```python
handle = Entrez.esummary(db="gene", id=list_of_genBank[0])
record = Entrez.read(handle, validate = True)
```

```python
print(record)
```

```python
handle = Entrez.efetch(db="gene", id=list_of_genBank[0], retmode='xml')
print(handle.read())
```
